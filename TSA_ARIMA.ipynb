### (a) Zt − 0.5Zt−1 = at

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Parameters for the AR(1) model
phi = 0.5
sigma_a = 1
n_obs = 100

# Generate white noise series
np.random.seed(42)  # For reproducibility
a = np.random.normal(0, sigma_a, n_obs)

# Initialize the series
Z = np.zeros(n_obs)

# Simulate the AR(1) series
for t in range(1, n_obs):
    Z[t] = phi * Z[t-1] + a[t]

# Plot the simulated series
plt.figure(figsize=(12, 6))
plt.plot(Z, label='Simulated AR(1) Series')
plt.xlabel('Time')
plt.ylabel('Zt')
plt.title('Simulated AR(1) Series Zt - 0.5Zt-1 = at')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the sample ACF and PACF
acf = sm.tsa.acf(Z, nlags=20)
pacf, confint = sm.tsa.pacf(Z, nlags=20, alpha=0.05)

# Plot the theoretical ACF
theoretical_acf = [phi**k for k in range(21)]
plt.figure(figsize=(12, 6))
plt.stem(range(21), acf, markerfmt='o', use_line_collection=True, label='Sample ACF')
plt.plot(range(21), theoretical_acf, 'r-', label='Theoretical ACF (0.5^k)')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.title('Sample ACF and Theoretical ACF Zt - 0.5Zt-1 = at')
plt.legend()
plt.grid(True)
plt.show()

# Plot the PACF with confidence intervals
plt.figure(figsize=(12, 6))
plt.stem(range(21), pacf, use_line_collection=True)
plt.fill_between(range(21), confint[:, 0] - pacf, confint[:, 1] - pacf, color='r', alpha=0.2)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.title('Sample PACF of Zt - 0.5Zt-1 = at')
plt.grid(True)
plt.show()

# Print ACF and PACF values
print("ACF:", acf)
print("PACF:", pacf)
```

**ACF:**

The ACF for an AR(1) process should follow the theoretical decay
ϕ^k

However, negative values  occur in the estimated ACF due to the finite sample size and estimation errors, even though the true autocorrelation should be positive.

- The ACF has a signifacnt peak at lag = 1
- It follows an exponential decay. This indicates that the observations are becoming less correlated as lag increases

**PACF:**

For an AR(1) process, the PACF should show a significant peak at lag 1 and then be effectively zero (within statistical noise) for higher lags. However here, due to the finite sample size, the PACF might not drop to zero exactly after lag 1, but it should be small and not statistically significant for lags greater than 1.

- A significant spike at lag 1 followed by spikes close to 0 at the preceeding lags indicates the AR(1) nature of this process


These effects seen in the ACF and the PACF highlights the role sample size and noise.

##The ACF: ρk =​γ0 / γk ​

We have:

ρ0 ​= 1

ρ1 = γ1/γ0 = − 1.8 /2.69 ≈ − 0.669

ρ2 = γ2 / γ0 = 0.5 2.69 ≈ 0.185

ρk = 0 for k>2

###**(b) Finding the ACF using the autocovariance generating function**

θ0 = 1

θ1 = 1.2

θ2 = -0.5

###The autocovariance generating function for AR(2):
γ(B) = σa^2 (1-θ1B- θ2B)(1-θ1B^-1 - θ2B^-1)

Where we assume σa^2 = 1
###Therefore, the autocovariances of the MA(2) model are:

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters
n_obs = 100  # Number of observations
ma_params = [-1.2, 0.5]  # MA coefficients for the MA(2) process
ar_params = []  # No AR parameters for this MA(2) process
sigma_a = 1  # Standard deviation of the white noise process

# Simulate the MA(2) process
ar = np.array([1])  # AR parameters with leading 1 for statsmodels
ma = np.array([1] + ma_params)  # MA parameters with leading 1 for statsmodels
arma_process = sm.tsa.ArmaProcess(ar, ma)
Zt = arma_process.generate_sample(nsample=n_obs, scale=sigma_a)

# Plot the simulated series
plt.figure(figsize=(14, 6))
plt.plot(Zt, label='Simulated MA(2) Process')
plt.title('Simulated Time Series Zt = (1-1.2B+0.5B^2)at')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Calculate and plot ACF and PACF
acf = sm.tsa.acf(Zt, nlags=20)
pacf = sm.tsa.pacf(Zt, nlags=20)

# Plot ACF
plt.figure(figsize=(14, 6))
plt.subplot(121)
plt.stem(range(21), acf, use_line_collection=True)
plt.title('Sample ACF')
plt.xlabel('Lag')
plt.ylabel('ACF')

# Plot PACF
plt.subplot(122)
plt.stem(range(21), pacf, use_line_collection=True)
plt.title('Sample PACF')
plt.xlabel('Lag')
plt.ylabel('PACF')

plt.tight_layout()
plt.show()

# Print ACF and PACF values
print("ACF:", acf)
print("PACF:", pacf)
print(Zt)
```

**ACF:**
- PACF at lag 0 is always 1, representing the correlation of the series itself
- At lag 1: value of -0.6625 is consistent with the negative coefficient in the MA(2) process (-1.2)
- At lag 2: The value of 0.1624 reflects the impact of the second term in the MA(2) process (0.5)
- ACFs after lag 2 are close to zero indictaing the process is of MA(2) nature

**PACF:**
- Lag 1: The significant negative spike aligns with the negative coefficient in the MA(2) process
- Lag 2: value of -0.5107 is significant, indicating the effect of the second lag
- The PACF values beyond lag 2 are smaller compared to those of lag 1 and lag 2 and hover around zero (consistent with the expectation that the PACF of an MA(2) process should drop to zero after lag 2)

###**(a) + (b) Simulating and plotting a series of n = 100 observations from each of the following models:**



### (i) Model 1: (1-B)Zt = (1-0.6B)at

This is an ARIMA(0,1,1) or IMA(1,1) Model as it can be rewritten as:

 Zt - Zt-1 = at - 0.6at-1

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

np.random.seed(42)

# Model 1: (1-B)Zt = (1-0.6B)at
ar1_diff = np.array([1, -1])
ma1 = np.array([1, -0.6])
model1 = ArmaProcess(ar1_diff, ma1)
z1 = model1.generate_sample(nsample=100)

# Plot the series
plt.figure(figsize=(10, 4))
plt.plot(z1)
plt.title('Model 1: (1-B)Zt = (1-0.6B)at')
plt.show()

# Display first few observations
print("Model 1 first few observations:", z1[:5])

# ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(z1, lags=20, ax=axes[0])
axes[0].set_title('ACF of Model 1')
plot_pacf(z1, lags=20, ax=axes[1])
axes[1].set_title('PACF of Model 1')

plt.tight_layout()
plt.show()

# Differences
diff_z1 = np.diff(z1)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(diff_z1, lags=20, ax=axes[0])
axes[0].set_title('ACF of Differenced Model 1')
plot_pacf(diff_z1, lags=20, ax=axes[1])
axes[1].set_title('PACF of Differenced Model 1')

plt.tight_layout()
plt.show()
```

###**(c)(i) Studying the sample ACF and PACF of the non-stationary and the stationary model**

**In this model, p=0, d=1 and q=1:**
- d=1: represents first order differencing
- q=1: indicates first order moving average term with parameter = θ = 0.6

The original series is non-stationary due to the presence of the differencing term (conditions for stationarity: contant mean, constant variance, no seasonality)

The differenced series ΔZt = Zt - Zt-1 is applied to this series to acheive stationarity

**ACF:**
- For the non-stationary series, the ACF shows a slow decay
- For the differenced (stationary) series, the ACF should resemble the MA(1) model
- For the differenced series, there is a significant spike at lag = 1, and the following ACFs for lag > 1 are insignificant

**PACF:**
- For the non-stationary series, there is a significant spike at lags 1 and 2. This is consistent with non-stationarity
- For the differenced  series, the PACF shows a significant spike at lag 1 followed by no significant ACFs in the following lags




### (ii) Model 2: (1-B)Zt = 5 + (1-0.6B)at

This is an ARIMA(0,1,1) or IMA(1,1) Model. It uses the same AR and MA components as Model 1 but adds a constant term 5 to the simulated series.

This can be rewritten as: Zt - Zt-1 = 5 + at - 0.6at-1

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

np.random.seed(42)

# Model 2: (1-B)Zt = 5 + (1-0.6B)at
ar2_diff = np.array([1, -1])
ma2 = np.array([1, -0.6])
model2 = ArmaProcess(ar1_diff, ma1)
z2 = model2.generate_sample(nsample=100) + 5

# Plot the series
plt.figure(figsize=(10, 4))
plt.plot(z2)
plt.title('Model 2: (1-B)Zt = 5 + (1-0.6B)at')
plt.show()

# Display first few observations
print("Model 2 first few observations:", z2[:20])

# ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(z2, lags=20, ax=axes[0])
axes[0].set_title('ACF of Model 2')
plot_pacf(z2, lags=20, ax=axes[1])
axes[1].set_title('PACF of Model 2')

plt.tight_layout()
plt.show()

# Differences
diff_z2 = np.diff(z2)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(diff_z2, lags=20, ax=axes[0])
axes[0].set_title('ACF of Differenced Model 2')
plot_pacf(diff_z2, lags=20, ax=axes[1])
axes[1].set_title('PACF of Differenced Model 2')

plt.tight_layout()
plt.show()
```

###**(c)(i) Studying the sample ACF and PACF of the non-stationary and the stationary model**

**In this model, p=0, d=1 and q=1:**
- d=1: represents first order differencing
- q=1: indicates first order moving average term with parameter = θ = 0.6
- IMA(0,1,1) with drift
- The drift term (5) contributes to the trend. This trend is removed when differencing, changing the behavior of the ACF and PACF.

**ACF:**
- For the non-stationary series, the model includes a drift term (5), which adds a constant value at each time step
- Due to the drift and the MA component, the ACF decays slowly, exhibiting a non-stationary behavior.
- For the differenced series, the trend introduced by the drift is rmeoved
- Significant spike at lag 1, then decays quickly (as previously mentioned, this exhibits the characteristic of the MA(1) process)

**PACF:**
- For the non-stationary series, there are significant values in the first few lags (~3)
- No clear pattern, this may be due to MA component and differencing
- For the differenced series, there is a significant spike at lag 1  (presence of MA(1))
- No significant spikes at higher lags (most lie within the bands or are very close to it)

### (iii) Model 3: (1-0.9B)(1-B)Zt = at

This is an ARIMA(1,1,0) Model.
This can be rewritten as: Zt - Zt-1 - 0.9(Zt-1 - Zt-2) = at

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

np.random.seed(42)

# Model 3: (1-0.9B)(1-B)Zt = at
ar3 = np.array([1, -1.9, 0.9]) # -1.9 due to expanding: (1−0.9B)(1−B)=1−B−0.9B+0.9B^2
ma3 = np.array([1])
model3 = ArmaProcess(ar3, ma3)
z3 = model3.generate_sample(nsample=100)

# Plot the series
plt.figure(figsize=(10, 4))
plt.plot(z3)
plt.title('Model 3: (1-0.9B)(1-B)Zt = at')
plt.show()

# Display first few observations
print("Model 3 first few observations:", z3[:20])

# ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(z3, lags=20, ax=axes[0])
axes[0].set_title('ACF of Model 3')
plot_pacf(z3, lags=20, ax=axes[1])
axes[1].set_title('PACF of Model 3')

plt.tight_layout()
plt.show()

# Differences
diff_z3 = np.diff(z3)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

plot_acf(diff_z3, lags=20, ax=axes[0])
axes[0].set_title('ACF of Differenced Model 3')
plot_pacf(diff_z3, lags=20, ax=axes[1])
axes[1].set_title('PACF of Differenced Model 3')

plt.tight_layout()
plt.show()
```

- d=1: represents first order differencing
- q=0: no MA term


**ACF:**
- For the non-stationary series, the ACF shows a slow decay => presence of non-stationarity
- Decay is also due to the AR(1) component (correlation between observations over time)
- For the differenced series, the ACF resembles that of an AR(1) Model
- Significant spike at lag 1, with clear gradual decline in the following lags

**PACF:**
- For the non-stationary series, there is a signifcant spike at lag 1, followed by smaller spikes within the bands
- For the differenced series, there is a signifcant spike at lag 1, followed by smaller spikes within the bands/ very close to the bands
- Empirical results might sometimes show some random fluctuations due to finite sample effects.

###**(c)(iv) Studying the sample ACF and PACF of the non-stationary and the stationary model**

**In this model, p=1, d=1 and q=1:**
- p=1: AR term of order 1
- d=1: represents first order differencing
- q=1: MA term of order 1


**ACF:**
- For the non-stationary series, the ACF shows a slow decay
- Due to the ARIMA(1,1,1) nature, we can see the effect of the AR(1), MA(1), and the differencing term

**PACF:**

### **The simulated series can differ due to variations in model implementation, despite the underlying process being theoretically the same. Therefore, there are multiple models for the four time series presented above and their simulated graphs will differ (although not greatly). However, the sample ACF and PACF of the stationary and non-stationary (differenced) series will exhibit similair characteristics and will prove that it indeed does belong to that certain ARIMA model (with respect to its specific p,d,q values)**

###Below are two examples to illustrating the proof of the statement above, these are based on Model 3 and Model 4. Unlike previous models, we will be utilizing the seaborn library for more aesthetically pleasing plot



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.arima_process import ArmaProcess

# Function to simulate ARIMA(1,1,0)
def simulate_arima_110(n):
    np.random.seed(43)
    ar = np.array([1, -0.9])
    ma = np.array([1])
    arma_process = ArmaProcess(ar, ma)
    return arma_process.generate_sample(nsample=n).cumsum()

# Function to simulate ARIMA(1,1,1)
def simulate_arima_111(n):
    np.random.seed(43)
    ar = np.array([1, -0.9])
    ma = np.array([1, -0.5])
    arma_process = ArmaProcess(ar, ma)
    return arma_process.generate_sample(nsample=n).cumsum()

# Simulate 100 observations
n = 100
arima_110_series = simulate_arima_110(n)
arima_111_series = simulate_arima_111(n)

print("Model 3 first few observations:", arima_110_series[:20])
print("Model 4 first few observations:", arima_111_series[:20])

# Plot the series
sns.set(style="darkgrid")
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
sns.lineplot(data=arima_110_series)
plt.title('Model 3 Series')

plt.subplot(2, 1, 2)
sns.lineplot(data=arima_111_series)
plt.title('Model 4 Series')

plt.tight_layout()
plt.show()

# Function to plot ACF and PACF
def plot_acf_pacf(series, lags=20):
    fig, ax = plt.subplots(2, 1, figsize=(14, 8))
    sm.graphics.tsa.plot_acf(series, lags=lags, ax=ax[0])
    sm.graphics.tsa.plot_pacf(series, lags=lags, ax=ax[1])
    plt.show()

# Plot ACF and PACF for both series
print("ACF and PACF for Model 3 Series")
plot_acf_pacf(arima_110_series)

print("ACF and PACF for Model 4 Series")
plot_acf_pacf(arima_111_series)

# Differencing to achieve stationarity
diff_arima_110_series = np.diff(arima_110_series, n=1)
diff_arima_111_series = np.diff(arima_111_series, n=1)

# Plot the differenced series
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
sns.lineplot(data=diff_arima_110_series)
plt.title('Differenced Model 3 Series')

plt.subplot(2, 1, 2)
sns.lineplot(data=diff_arima_111_series)
plt.title('Differenced Model 4 Series')

plt.tight_layout()
plt.show()

# Plot ACF and PACF for differenced series
print("ACF and PACF for Differenced Model 3 Series")
plot_acf_pacf(diff_arima_110_series)

print("ACF and PACF for Differenced Model 4 Series")
plot_acf_pacf(diff_arima_111_series)
```
